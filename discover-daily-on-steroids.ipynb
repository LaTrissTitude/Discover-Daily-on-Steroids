# To visualize music, we first need to create the universe

Music is cultural, produced by artists somewhere on the planet, at some point in time. 
With this dataset, we've got each track's publishing date. Let's identify where it may have came from !

For now, let's try to predict countries from track names.

```python
import pandas as pd
import numpy as np
tracks = pd.read_csv("/kaggle/input/spotify-dataset-19212020-600k-tracks/tracks.csv")

names = tracks["name"]


```

```python
import fasttext
model_url = "https://dl.fbaipublicfiles.com/fasttext/supervised-models/lid.176.ftz"
fmodel = fasttext.load_model(model_url)
def detect_lang(txt):
    return fmodel.predict(txt)
test = names.head(10)
result = test.apply(detect_lang)
print(pd.DataFrame(zip(test, result)))
```
